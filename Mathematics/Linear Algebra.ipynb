# Linear Algebra

## Vectors

Vectors have <b>magnitude</b> and <b>direction</b> <br><br>
So what is an example of a vector? <br>
- <b>Speed</b> cannot be a vector because it only has <b>magnitude</b>. In linear algebra, this is called a <b>scalar</b>.
<i>e.g. speed is 5mph</i><br>
- <b>Velocity</b>, however, is a <b>vector</b> because it has <b>magnitude</b> and <b>direction</b>.
<i>e.g. velocity is 5mph going <u>East</u></i><br><br>
Notation:
$\vec{V}=(3, 2)=\Big[\begin{array}{rcl}3\\2\end{array}\Big]$ <br><br>
This means we are moving 3 units in the positive, horizontal direction<br>
But moving 2 units in the positive, vertical direction

```python
# Run this to show a graph of the vector

%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np

plt.rc('font', size=15)
plt.arrow(0, 0, 3, 2, length_includes_head=True,
          head_width=0.2, head_length=0.2)
plt.text(1.5, 1.5, r"$\vec{V}$")

plt.axis([0, 5, 0, 5])
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title(r'Graph showing vector  $\vec{V}$')
plt.show()
```

## Vector addition

If we have two vectors:<br><br>
$\vec{a}=\Big[\begin{array}{rcl}6\\{-2}\end{array}\Big]$ <br>
$\vec{b}=\Big[\begin{array}{rcl}{-4}\\4\end{array}\Big]$ <br><br>
Then we add them together, we get:<br><br>
$\vec{a}+\vec{b}=\Big[\begin{array}{rcl}6\\{-2}\end{array}\Big]+\Big[\begin{array}{rcl}{-4}\\{4}\end{array}\Big]=\Big[\begin{array}{rcl}{2}\\{2}\end{array}\Big]$ 

```python
# Run this to show a graph of vector addition

%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np

plt.rc('font', size=15)

# Vector A
plt.arrow(0, 0, 6, -2, length_includes_head=True,
          head_width=0.2, head_length=0.2)
plt.text(4, -3, r"$\vec{A}$")

# Vector B
plt.arrow(0, 0, -4, 4, length_includes_head=True,
          head_width=0.2, head_length=0.2)
plt.text(-3, 1.5, r"$\vec{B}$")

# Vector A+B
plt.arrow(0, 0, 3, 2, length_includes_head=True,
          head_width=0.2, head_length=0.2)
plt.text(2.5, 2.5, r"$\vec{A}+\vec{B}$")

# Setting labels
plt.axis([-5, 7, -4, 5])
plt.grid(True)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title(r'Graph showing vector  $\vec{A}+\vec{B}$')
plt.show()
```

## Multiplying a vector by a scalar

If we have a vector $\vec{A}$:<br>
$\vec{A}=\Big[\begin{array}{rcl}2\\1\end{array}\Big]$<br><br>
and we multiply it with the scalar 3, we get:<br>
$3\vec{A}=3\Big[\begin{array}{rcl}2\\1\end{array}\Big]=
\Big[\begin{array}{rcl}3\cdot{2}\\3\cdot{1}\end{array}\Big]=
\Big[\begin{array}{rcl}6\\3\end{array}\Big]$<br><br>
<i>Note: </i>
' $\cdot$ '
<i>means multiply in mathematics</i>
